```python
import pandas as pd
from sklearn.preprocessing import *
```

```python
# getting data
df = pd.read_csv("ai_job_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# check for duplicate values
df.duplicated().sum()
```

```python
# check for null values
df.isnull().sum()
# there are no missing values 
```

```python
#droping unneeded columns
df.drop(['salary_currency','job_id', 'company_name', 'posting_date', 'application_deadline', 'required_skills'], axis=1, inplace=True)
```

```python
df.info()
```

Checking all the data 

```python
# job title
df["job_title"].unique()
```

```python
one_hot_encoder = OneHotEncoder( )
job_title_one_hot_encoded = one_hot_encoder.fit_transform(df[["job_title"]])
feature_names_job_title = one_hot_encoder.get_feature_names_out(["job_title"])
encoded_job_title_df = pd.DataFrame(job_title_one_hot_encoded.toarray(), columns=feature_names_job_title, index=df.index) 

# drop the job_title from the table 
df = df.drop('job_title', axis=1)

# concat the table 
df = pd.concat([df, encoded_job_title_df], axis=1)
```

```python
df.info()
```

```python
# experience_level 
df["experience_level"].unique()
```

```python
# apply one hot ending over the "experence_level"
experience_level_encoded = one_hot_encoder.fit_transform(df[["experience_level"]])
feature_name_experience_level = one_hot_encoder.get_feature_names_out(["experience_level"])
encoded_experience_level_df = pd.DataFrame(experience_level_encoded.toarray(),columns=feature_name_experience_level)

# contact 
pd.concat([df,encoded_experience_level_df],axis=1)
#removeing 
df = df.drop("experience_level",axis=1)
```

```python
df.head()
```

```python
#employment_type   
df["employment_type"].unique()
```

```python
# one hot encoding over the employment_type  
employment_type_encoded = one_hot_encoder.fit_transform(df[["employment_type"]])
feature_name_employment_type  = one_hot_encoder.get_feature_names_out(["employment_type"])
encoded_employment_type_df = pd.DataFrame(employment_type_encoded.toarray(),columns=feature_name_employment_type)

# removing 
df = df.drop("employment_type",axis=1)
#concat
df = pd.concat([df,encoded_employment_type_df],axis=1)
df.head()
```

```python
# company_location 
df["company_location"].unique()
```

```python
company_location_encoded = one_hot_encoder.fit_transform(df[["company_location"]])
feature_name_company_location  = one_hot_encoder.get_feature_names_out(["company_location"])
company_location_encoded_df = pd.DataFrame(company_location_encoded.toarray(),columns=feature_name_company_location)

#removing 
df = df.drop("company_location",axis=1)

#concat
df = pd.concat([df,company_location_encoded_df],axis=1)
df.head()
```

```python
# company_size
df["company_size"].unique()
```

```python
# one hot encoding over company_size
encoded_company_size = one_hot_encoder.fit_transform(df[["company_size"]])
feature_name_company_size = one_hot_encoder.get_feature_names_out(["company_size"])
encoded_df_company_size_df = pd.DataFrame(encoded_company_size.toarray(), columns=feature_name_company_size)

# removing 
df = df.drop("company_size",axis=1)

#concat
df = pd.concat([df,encoded_df_company_size_df],axis=1)
df.head()
```

```python
# employee_residence      
df["employee_residence"].unique()
```

```python
# one hot encoding over employee_residence
encoded_employee_residence = one_hot_encoder.fit_transform(df[["employee_residence"]])
feature_name_employee_residence = one_hot_encoder.get_feature_names_out(["employee_residence"])
encoded_df_employee_residence_df = pd.DataFrame(encoded_employee_residence.toarray(), columns=feature_name_employee_residence)

# removing 
df = df.drop("employee_residence",axis=1)

#concat
df = pd.concat([df,encoded_df_employee_residence_df],axis=1)
df.head()
```
